```python
import pandas as pd
import numpy as np
```

```python
planes = pd.read_csv('planes_fil.txt', encoding='utf-8')
```

```python
planes = planes.rename(index=str, columns={'Unnamed: 0': 'idx'})
planes.iloc[1].idx
```

```python
letters = ['A', 'B', 'C', 'D', 'E', 'F']
def seattype(i, pseats):
    types = ['ekonomiczna', 'biznes', 'premium']
    if (i < pseats*0.01):
        return types[2]
    elif (i < pseats*0.05):
        return types[1]
    else:
        return types[0]
```

```python
result = np.empty((0, 3))
for i in range(planes.shape[0]):
    seatsLet = (letters[x/50] for x in np.arange(0, planes.iloc[i]['Seats'] ))
    seatsNum = (x%50+1 for x in np.arange(0, planes.iloc[i]['Seats'] ))
    plane = planes.iloc[i]
    for j in range(plane.Seats):
        num = seatsNum.next()
        stype = seattype(num, plane.Seats)
        result = np.vstack((result, np.array([plane.idx, seatsLet.next() + str(num), stype])))
```

```python
result
```

```python
df = pd.DataFrame(data=result, columns=['Model', 'Seat', 'Class'])
```
