```python
import pandas as pd
import numpy as np
```

```python
models = pd.read_csv('planes_fil.txt', encoding='utf-8')
models['Unnamed: 0']
```

```python
planes_num = 500;
model = np.random.choice(models['Unnamed: 0'], planes_num)
name = [random_namepair() for x in range(planes_num)]
operates = np.random.choice(['t', 'f'], planes_num, p = [0.985, 0.015])
```

```python
d = {'Name': name, 'Model': model, 'Operates': operates}
df = pd.DataFrame(data=d)
```

```python
df.index += 1
df
```

```python
df[['Name', 'Model', 'Operates']].to_csv('aircraft_fil.txt', index=True, encoding='utf-8');
```
