```python
import pandas as pd
import numpy as np
```

```python
#airports = pd.read_csv('airports_fil.txt', encoding='utf-8')
#airports[airports.duplicated('City')]
fresh = pd.read_csv('airports.dat', encoding='utf-8', header=None)
fresh.columns = [
'AirportID',
'Name',
'City',
'Country',
'IATA',
'ICAO',
'Latitude',
'Longitude',
'Altitude',
'Timezone',
'DST',
'Tz database time zone',
'Type',
'Source']
countries = pd.read_csv('countries_fil.txt', encoding='utf-8')
```

```python
df = pd.merge(fresh, countries, left_on='Country', right_on='Name')
```

```python
df.index += 1
```

```python
filtered = df[(df.City != '\N') & (pd.notnull(df.City))
              & (df.Country != '\N') & (pd.notnull(df.Country))]
filtered.City = filtered.City.apply(lambda x: x.replace(',','.'))
filtered[['City', 'Country', 'iso_code']]
```

```python
#filtered[filtered.duplicated('City')]
filtered.drop_duplicates('City', inplace = True)
filtered[filtered.duplicated('City')]
```

```python
filtered[['City', 'iso_code']].to_csv('cities_fil.txt', index=True, encoding='utf-8')
```
