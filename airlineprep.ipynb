```python
import pandas as pd
```

```python
df = pd.read_csv('airlines.txt', encoding='utf-8')
```

```python
df.head()
```

```python
df.columns = ['AirlineID',
'Name',
'Alias',
'IATA',
'ICAO',
'Callsign',
'Country',
'Active']
```

```python
df[['Name', 'Alias']].head()
```

```python
filtered = df[(df.Active == 'Y') & (df.IATA != '\N') & (pd.notnull(df.IATA)) & (df.ICAO != '\N') & (pd.notnull(df.ICAO))]
filtered = filtered[(df.AirlineID != 13391) & (df.AirlineID != 13394) & (df.AirlineID != 19567)]
```

```python
filtered[['AirlineID', 'Name', 'IATA', 'ICAO', 'Country']]
```

```python
filtered[['AirlineID', 'Name', 'IATA', 'ICAO', 'Country']].to_csv('airlines_fil.txt', index=False, encoding='utf-8');
```
